```python
import torch
from opt_src import config_parser
from data import dataset_dict
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
scene = 'lego'
ckpt = 'base-3src-dense.tar'
num_src_views = 3
```

```python
cmd = f'--datadir /mnt/hdd/mvsnerf_data/nerf_synthetic/{scene}  \
         --dataset_name blender_src --white_bkgd \
        --net_type v0 --ckpt ./ckpts/{ckpt} --num_src_views {num_src_views}'

args = config_parser(cmd.split())
args.use_viewdirs = True

args.N_samples = 128
args.feat_dim =  8+12
```

```python
dataset_val = dataset_dict['blender_src'](args, split='val', select_index=None)
```

```python
dataset_train = dataset_dict['blender_src'](args, split='train', select_index=None)
```

```python
dataset_test = dataset_dict['blender_src'](args, split='test', select_index=None)
```
